```python
import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
pd.isnull(ufo).head()
```

```python
ufo.isnull().head()
```

```python
ufo.loc[0:4, :]
```

```python
ufo.iloc[0:4, :]
```

```python
# 为什么loc和iloc有以下的差别：
# pd.loc[0:4, :] 包含0, 1, 2, 3, 4
# pd.iloc[0:4, :] 包含0, 1, 2, 3
```

```python
# iloc继承自numpy
ufo.values[0:4, :]
```

```python
list(range(0, 4))
```

```python
# random sampling
ufo.sample(n = 3)
```

```python
# 设置随机数种子，每次采样得到相同样本
ufo.sample(n = 3, random_state = 0)
```

```python
ufo.sample(frac = 0.1, random_state = 0)
```

```python
train = ufo.sample(frac = 0.75, random_state = 0)
```

```python
test = ufo.loc[~ufo.index isin (train.index), :]
```
