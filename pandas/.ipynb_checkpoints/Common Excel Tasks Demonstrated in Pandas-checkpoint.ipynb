```python
import pandas as pd
import numpy as np
df = pd.read_excel('excel-comp-data.xlsx')
```

```python
df.head()
```

```python
# 1.add a sum to a row
df['Total'] = df['Jan'] + df['Feb'] + df['Mar']
df.head()
```

```python
# 2.在底部加上合计行
# 2.1 求和各个数据列
sum_row = df[['Jan', 'Feb', 'Mar', 'Total']].sum()
sum_row
```

```python
# 2.2 把结果转置成数据集
df_sum = pd.DataFrame(data = sum_row).T
df_sum
```

```python
# 2.3 加上其他变量
df_sum = df_sum.reindex(columns = df.columns)
df_sum
```

```python
# 2.4 把汇总结果加到最后一行
df.final = df.append(df_sum, ignore_index=True)
df.final
```
