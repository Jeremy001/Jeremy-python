## Video_04.figure图像

```python
# 画两个图

## 生成随机数
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = x ** 2

## 绘制第一个图
plt.figure()
plt.plot(x, y1)

## 绘制第二个图
plt.figure()
plt.plot(x, y2)

## 展示图形
plt.show()
```

```python
# 进行设置

## 第一个图
plt.figure()
plt.plot(x, y1)

## 第二个图，设置图形编号为3， 图形大小设置为10：6
plt.figure(num=3, figsize=(10, 6))
plt.plot(x, y1)
### 在图2中绘制两条线，并设置第二条线的颜色，线宽和线型
plt.plot(x, y2, color = 'red', linewidth = 1.0, linestyle = '--')

## 展示两个图形
plt.show()
```

## Video_05.设置坐标轴(1)

```python
# 生成数据
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = x ** 2

# 绘制图形
plt.figure(figsize = (10, 8))
plt.plot(x, y1)
plt.plot(x, y2, color = 'red', linewidth = 1.0, linestyle = '--')
## 设置坐标轴的范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))
## 设置坐标轴标题
plt.xlabel('I am x')
plt.ylabel('I am y')
## 设置x轴ticks
new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
## 设置y轴ticks，用文字来表示
plt.yticks([-2, -1, 0, 1, 2], 
           ['really bad', 'bad', 'normal', 'good', 'really good'])
# 展示图形
plt.show()
```

## Video_06.设置坐标轴(2)

```python
# 生成数据
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = x ** 2

# 绘制图形
plt.figure(figsize = (10, 8))
plt.plot(x, y1)
plt.plot(x, y2, color = 'red', linewidth = 1.0, linestyle = '--')
## 设置坐标轴的范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))
## 设置坐标轴标题
plt.xlabel('I am x')
plt.ylabel('I am y')
## 设置x轴ticks
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
## 设置y轴ticks，用文字来表示
plt.yticks([-2, -1, 0, 1, 2], 
           ['really bad', 'bad', 'normal', 'good', 'really good'])
## 移动spines
### gca = 'get current axis'
ax = plt.gca()
### 设置上边和右边为空
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
### 绑定x轴为底边，y轴为左边
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
### 移动x轴
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# 展示图形
plt.show()
```
