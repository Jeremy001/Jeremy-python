## 1. Controlling figure aesthetics
***

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(sum(map(ord, 'aesthetics')))
```

```python
# define a simple function to plot some offset sine waves
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
# plot
sinplot()
```

```python
#  switch to seaborn defaults, simply call the set() function.
sns.set()
sinplot()
```

```python
# There are five preset seaborn themes: darkgrid, whitegrid, dark, white, and ticks
# The default theme is darkgrid. 
sns.set_style('darkgrid')
sinplot()
```

```python
sns.set_style('whitegrid')
sinplot()
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
# 有横纵坐标刻度线
sns.set_style('ticks')
sinplot()
```

```python
sns.set_style('whitegrid')
data = np.random.normal(size = (20, 6)) + np.arange(6) / 2
sns.boxplot(data = data)
```

```python
# 去掉顶部和右侧的边框
# 只有theme = white or ticks才可以去掉
sns.set_style('ticks')
sinplot()
sns.despine()
```

```python
# offsetting the spines 
f, ax = plt.subplots()
sns.violinplot(data = data)
sns.despine(offset = 10, trim = True)
```

```python
# define which spines are removed
sns.set_style('whitegrid')
sns.boxplot(data = data, palette='deep')
sns.despine(left = True)
```

```python
# 为某个图形单独设置样式 with
# in a with statement to temporarily set plot parameters
with sns.axes_style('whitegrid'):
    plt.subplot(221)
    sinplot()
with sns.axes_style('darkgrid'):
    plt.subplot(222)
    sinplot()
plt.subplot(223)
sinplot()
plt.subplot(224)
sinplot()
# plt.subplot(ABC):A行数 B列数 C第几个子图，第一行从左往右数
```

#### 2. Choosing color palettes

- color_palette()

- husl_palette()

```python
sns.set(rc = {'figure.figsize':(20, 14)})
np.random.seed(sum(map(ord, 'palettes')))
```

```python
# 当前palette
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
# 设置想要展现的颜色类别数
# 默认是6个，当多于6个时，可以自定义颜色数
sns.palplot(sns.color_palette('hls', 10))
# 以上默认的明度和饱和度是0
# 也可以自定义明度和饱和度
sns.palplot(sns.hls_palette(10, l = .3, s = .8))
sns.palplot(sns.hls_palette(10, l = .5, s = .8))
```

```python
# 由于人类视觉系统的工作性质，即使同样强度的颜色，我们认为黄和绿更亮而蓝色更暗
# 因此seaborn有另外一套更标准的颜色系统husl, 产生得到的颜色让人看起来亮度是一样的
sns.palplot(sns.color_palette('husl', 10))
sns.palplot(sns.husl_palette(10, l = .5, s = .8))
```

### 3.1 Plotting univariate distributions

```python
# 准备工作：加载所需的类和进行必要的设置
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes = True)
np.random.seed(sum(map(ord, 'distribution')))
```

```python
# 绘制直方图的最快捷方式：distplot()
## 默认显示核密度曲线
x = np.random.normal(size = 100)
sns.distplot(x)
```

```python
# 去除核密度曲线，显示rug
sns.distplot(x, kde = False, rug = True)
```

```python
# 修改bins的数量
## 通过修改bins数量，可以探索该变量的分布
sns.distplot(x, bins = 20, kde = False, rug = True)
```

```python
# 核密度估计曲线
sns.distplot(x, hist=False, rug = True)
```

```python
# 绘制核密度估计曲线的更快捷方式
sns.kdeplot(x, shade = True)
```

```python
# 修改bandwidth
sns.kdeplot(x)
sns.kdeplot(x, bw = .2, label = 'bw:.2')
sns.kdeplot(x, bw = 2, label = 'bw:2')
plt.legend()
```

```python
# cut
sns.kdeplot(x, shade = False, cut = 0)
sns.rugplot(x)
```

```python
# Fitting parametric distributions
x = np.random.gamma(6, size = 200)
sns.distplot(x, kde = False, fit = stats.gamma)
```

### 3.2 Plotting bivariate distributions

```python
# 散点图
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=['x', 'y'])
sns.jointplot(x = 'x', y = 'y', data = df)
```

```python
# 六边形图
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style('white'):
    sns.jointplot(x = x, y = y, kind = 'hex', color = 'k')
```

```python
# 核密度估计曲线
sns.jointplot(x, y, data = df, kind = 'kde')
```

### 3.3 Visualizing pairwise relationships in a dataset

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
# 更灵活的设置
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, cmap = 'Blues_d', n_levels = 6)
```

### 4.1 Categorical scatterplots

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips)
# stripplot中的点可能会重叠，不太容易看出data的真实分布
```

```python
# 给stripplot添加jitter
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True)
```

```python
# 另外一个解决数据点重叠问题的方法是使用swarmplot
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)
```

```python
# 可以在swarmplot中添加一个分类变量，区分数据点
# sns.stripplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', jitter = True)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
# 在swarmplot中，X轴分类变量会根据出现顺序进行排序，如果分类变量是数字形式，那么会根据大小排序
sns.swarmplot(x = 'size', y = 'total_bill', data = tips)
```

```python
# X轴也可以放置数值型变量，y轴也可以放分类型变量
sns.swarmplot(x = 'total_bill', y = 'day', hue = 'time', data = tips)
# 从周五开始，大家基本都吃晚饭，而周四多数是午饭
```
