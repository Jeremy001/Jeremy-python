## Video_06.设置坐标轴(2)

```python
# 生成数据
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = x ** 2

# 绘制图形
plt.figure(figsize = (10, 8))
plt.plot(x, y1)
plt.plot(x, y2, color = 'red', linewidth = 1.0, linestyle = '--')
## 设置坐标轴的范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))
## 设置坐标轴标题
plt.xlabel('I am x')
plt.ylabel('I am y')
## 设置x轴ticks
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
## 设置y轴ticks，用文字来表示
plt.yticks([-2, -1, 0, 1, 2], 
           ['really bad', 'bad', 'normal', 'good', 'really good'])
## 移动spines
### gca = 'get current axis'
ax = plt.gca()
### 设置上边和右边为空
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
### 绑定x轴为底边，y轴为左边
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
### 移动x轴
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# 展示图形
plt.show()
```

## Video_07.设置图例

```python
plt.figure(figsize = (10, 8))
plt.xlim(-1, 2)
plt.ylim(-2, 3)
plt.xlabel('I am x')
plt.ylabel('I am y')
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks([-2, -1, 0, 1, 2], 
           ['really bad', 'bad', 'normal', 'good', 'really good'])
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
# 设置label，图例中展示的就是label的值
# 设置l1, l2, 为了在plt.legend()中是指handles和labels参数
l1, = plt.plot(x, y1, label = 'up')
l2, = plt.plot(x, y2, color = 'red', linewidth = 1., linestyle = '--', label = 'down')
# handles:设置图例中展示哪些条曲线
# labels:设置新的label，不使用上面设置的label
# loc:设置图例的位置，可取值有:
# center,upper left,right, lower center, lower right, upper center, center left, best, upper right, lower left, center right等
plt.legend(handles = [l1, l2,], 
           labels = ['a', 'b'], 
           loc = 'center right')
plt.show()
```

## Video_08.Annotation 标注

```python
x = np.linspace(-3, 3, 50)
y = 2 * x + 1

plt.figure(figsize = (10, 8))
plt.plot(x, y, color = 'blue')
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# 添加点
x0 = 1
y0 = 2 * x0 + 1
# s = 50, 点的大小
plt.scatter(x0, y0, s = 80, color = 'red')

## 添加线
plt.plot([x0, x0], [y0, 0], color = 'k', linestyle = '--', linewidth = 3)

## 添加标注(1)
plt.annotate(r'$2x+1=%s$' % y0,   # r表示正则，文本用$$括起来，%s表示传入一个参数， 最后的% y0表示把y0传入到文本中
             xy = (x0, y0),   # 表示以(x0, y0)点开始
             xycoords = 'data', 
             xytext = (+30, -30),   # 表示在(x0 + 30, y0-30)的位置开始显示文本标注
             textcoords = 'offset points',   # 表示坐标的方式是offset points，即在某点的offset后的点
             fontsize = 16, 
             arrowprops = dict(arrowstyle = '->',   # 箭头的样式
                               connectionstyle = 'arc3, rad = .2'))  # 箭头的弧度
# 添加标注(2)
plt.text(-3, 3,   # 标注开始的位置
         r'$This\ is\ the\ annotation.\ \mu.\ \sigma_i.\ \alpha_i.$', 
         fontdict = {'size':16, 'color':'r'})  # 对字体的设置，一个字典

plt.show()
```

## Video_11.柱形图bar

```python
# 生成随机数字
n = 12
X = np.arange(12)
y1 = (1- X / float(12)) * np.random.uniform(0.5, 1.0, n)
y2 = (1- X / float(12)) * np.random.uniform(0.5, 1.0, n)

# 绘制图形
plt.figure(figsize = (10, 8))
## 设置柱子的颜色
plt.bar(X, +y1, facecolor = '#9999ff', edgecolor = 'white')
plt.bar(X, -y2, facecolor = '#ff9999', edgecolor = 'white')

# 添加数据标签
for x, y in zip(X, y1):
    # ha:horizontal alignment
    # va:vertical alignment
    plt.text(x, y + 0.04, '%.2f' % y, ha = 'center', va = 'bottom')

for x, y in zip(X, y2):
    plt.text(x, -y-0.04, '-%.2f' % y, ha = 'center', va = 'top')
    
# 设置坐标轴为空
plt.xlim(-1, n)
plt.ylim(-1.25, 1.25)
plt.xticks(())
plt.yticks(())

# 设置四边颜色为蓝色
ax = plt.gca()
ax.spines['top'].set_color('b')
ax.spines['bottom'].set_color('b')
ax.spines['left'].set_color('b')
ax.spines['right'].set_color('b')

plt.show()
```

## Video_11.等高线图contour

```python
# 定义计算高度的公式
def f(x, y):
    return (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)

# 生成想x, y
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
## 把数据转成坐标轴矩阵
X, Y = np.meshgrid(x, y)

plt.figure(figsize = (10, 8))
# 填充等高线图的颜色
plt.contourf(X, Y, f(X, Y),   # 点
             8,    # 设置等高线图分成多少个等高区域，数字越大越密集
             alpha = .75, 
             cmap = plt.cm.hot)   # 设置cmap的样式，hot表示暖色

# 添加等高线
C = plt.contour(X, Y, f(X, Y), 8, colors = 'black', linewidth = .5)

# 添加label
plt.clabel(C, inline = True, fontsize = 10)

# 去掉xticks和yticks
plt.xticks(())
plt.yticks(())

# 展示图形
plt.show()
```

## Video_14.3D图

```python
# load class
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 生成数据
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X **2 + Y **2)
Z = np.sin(R)

# 绘制图形
fig = plt.figure(figsize = (10, 8))
## 添加3d坐标
ax = Axes3D(fig)
## 绘制3d图
ax.plot_surface(X, Y, Z,    # 点的坐标
                rstride = 1, cstride = 1,    # 设置图形的密集度，数据越大越稀疏
                cmap = plt.get_cmap('rainbow'))   # 设置图形的颜色
# 添加等高线图
ax.contourf(X, Y, Z, zdir = 'z', offset = -2, cmap = 'rainbow')
ax.set_zlim(-2, 2)
plt.show()
```

## Video_16.子图，分格显示

```python
# 方法1：subplot2grid
plt.figure(figsize = (12, 10))
## 子图1，第一行，占据3列
ax1 = plt.subplot2grid((3, 3),   # 表示整个图形分3行3列
                       (0, 0),   # 表示第一个子图是第一行第一列
                       colspan = 3,   # 表示子图占据3列的位置
                       rowspan = 1)   # 表示子图占据一行的位置
ax1.plot([0, 2], [0,2])
ax1.set_title('ax1 title')

## 子图2， 第二行，占据2列
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan = 2)
ax2.plot([0, 1], [0, 1])
ax2.set_title('ax2 title')

## 子图3， 第三列，但是占据第二第三行
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan = 2)
ax3.plot([0, 1], [0, 3])
ax3.set_title('ax3 title')

## 子图4， 第三行，第一列
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax4.plot([0, 1], [0, 1])
ax4.set_title('ax4 title')
## 子图5， 第三行，第二列
ax5= plt.subplot2grid((3, 3), (2, 1))
ax5.plot([0, 1], [0, 1])
ax5.set_title('ax5 title')

# 显示图形
plt.show()
```

```python
# 方法2：gridspec
import matplotlib.gridspec as gridspec

plt.figure()
## 定义一个3行3列的grid 
gs = gridspec.GridSpec(3, 3)

ax1 = plt.subplot(gs[0, :])
ax1.plot([0, 1], [0, 4])

ax2 = plt.subplot(gs[1, :2])
ax2.plot([0, 1], [0, 1])

ax3 = plt.subplot(gs[1:, 2])
ax3.plot([0, 1], [0, 4])

ax4 = plt.subplot(gs[2, 0])
ax4.plot([0, 1], [0, 1])

ax5 = plt.subplot(gs[2, 1])
ax5.plot([0, 1], [0, 1])
plt.show()
```

## Video_17.图中图

## Video_19.动画animation

```python
# 导入类
from matplotlib import animation

# 绘制动画
fig, ax = plt.subplots()
# 生成数据
x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))

# 定义动画function
def animate(i):
    line.set_ydata(np.sin(x + i/100))
    return line,

# 定义init_function
def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig = fig, 
                              func = animate, 
                              frames = 100, 
                              init_func = init, 
                              interval = 20, 
                              blit = False)
plt.show()

# 但是jupyter notebook上貌似看不了动画
```
