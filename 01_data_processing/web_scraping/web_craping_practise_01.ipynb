```python
# load packages
import requests
from bs4 import BeautifulSoup
```

```python
# requests.get()的作用是下载页面
page1 = requests.get("http://www.chinasspp.com/brand/服装品牌/1/")
# 查看页面是否顺利下载，当返回的code是200时，表示顺利下载了
page1
# 如果code以4或5开头，一般表示报错了，未顺利下载
```

```python
# 查看页面内容
page1.content
```

```python
# 用BeautifulSoup解析页面
# from_encoding = 'gb18030'的作用是使中文正确显示，不会显示成乱码
soup1 = BeautifulSoup(page1.content, 'html.parser', from_encoding='gb18030')
```

```python
# 用prettify美化，然后打印出来
print(soup1.prettify())
```

```python
# 查找标签，指定class
# 注意，指定class时写的是class_,末尾有"_"符号
soup1.find_all('div', class_ = 'brand')
# 公司的信息放在div中，class = 'brand'
```

```python
soup1.select('div.brand p span')
```
