```python
# load packages
from requests_html import HTMLSession
import pandas as pd
# 目标描述：
# 把页面上的每一个课程标题和对应链接爬取下来，保存到excel表格中
```

```python
# 建立一个回话session
session = HTMLSession()
```

```python
# 开始工作
## 设置url
url = 'https://www.jianshu.com/p/85f4624485b9'
## 利用session的get功能把网页取到
webpage = session.get(url)
```

```python
# 打印页面的文本
print(webpage.html.text)
```

```python
# 页面中的所有链接
webpage.html.links
# 发现很多链接不是我们想要的，并不是课程的链接， 比如sign_in,sign_up
# 原因是：这些链接都是相对链接，而我们课程对应的链接都是绝对链接
```

```python
# 绝对链接
webpage.html.absolute_links
# 但是我们也发现，下面的链接中有些也不是我们想要的
```

```python
# 通过检查源代码，复制出前面几个链接的selector:
# body > div.note > div.post > div.article > div.show-content > div > p:nth-child(4) > a  
# body > div.note > div.post > div.article > div.show-content > div > p:nth-child(6) > a  
# body > div.note > div.post > div.article > div.show-content > div > p:nth-child(9) > a
```

```python
# 我们可以根据一个链接的selector来抓取该链接
sel = 'body > div.note > div.post > div.article > div.show-content > div > p:nth-child(4) > a'
results = webpage.html.find(sel)
print('链接：')
print(results)
# 发现取到了链接，但是没有链接对应的文本
# 文本如下：
print('文本：')
print(results[0].text)
# 链接主体
print('链接主体：')
print(results[0].absolute_links)
# 链接主体，取文本内容
print('文本链接：')
print(list(results[0].absolute_links)[0])
```

```python
# 编写函数，把文本和链接一次性返回
def get_text_link(sel):
    mylist = []
    try:
        results = webpage.html.find(sel)
        for result in results:
            mytext = result.text
            mylink = list(result.absolute_links)[0]
            mylist.append((mytext, mylink))
        return mylist
    except:
        return none
```

```python
# 测试一下编写的函数
print(get_text_link(sel))
# 发现函数可用
```

```python
# 利用get_text_link函数批量爬取数据
# 注意先调整一下sel，观察sel发现，只要改成下面结构就可以了
sel = 'body > div.note > div.post > div.article > div.show-content > div > p > a'
text_link = get_text_link(sel)
print(text_link)
```

```python
df_text_link = pd.DataFrame(get_text_link(sel))
```

```python
df_text_link
```

```python
df_text_link.columns = ['text', 'link']
```

```python
df_text_link.head()
```

```python
# 保存到本地，存为csv文件
df_text_link.to_csv('text_link.csv', 
                    encoding = 'gbk', 
                   index = False)
```
