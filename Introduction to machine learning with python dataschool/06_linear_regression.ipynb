```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# 1.read csv file into python
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', 
                   index_col = 0)
# 展示数据集的前5行
data.head()
```

```python
# 查看数据集的大小
data.shape
```

1.features:

- TV: advertising dollars spent on TV for a single product in a given market (in thousands of dollars)
- Radio: advertising dollars spent on Radio
- Newspaper: advertising dollars spent on Newspaper

2.response:

- Sales: sales of a single product in a given market (in thousands of items)

```python
# 2.可视化探索
import seaborn as sns
%matplotlib inline
sns.pairplot(data, 
             x_vars = ['TV', 'radio', 'newspaper'], 
             y_vars = 'sales', 
             size = 7, 
             aspect = .7, 
             kind = 'reg')
```

```python
# 3.建立线性回归模型
### 3.1 select features and target
feature_cols = ['TV', 'radio', 'newspaper']
X = data[feature_cols]
y = data.sales
### 3.2 split into train and test set
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
#### defualt test_size = 25%
### 3.3 linear regression model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
print(linreg.intercept_)
print(linreg.coef_)
list(zip(feature_cols, linreg.coef_))
```

```python
# 4.预测
y_pred = linreg.predict(X_test)
```

5.模型评估

- Mean Absolute Error (MAE)
- Mean Squared Error (MSE) 
- Root Mean Squared Error (RMSE)

```python
# 5.模型评估
# 5.1 MAE
from sklearn import metrics
metrics.mean_absolute_error(y_test, y_pred)
```

```python
# 5.2 MSE
metrics.mean_squared_error(y_test, y_pred)
```

```python
# 5.3 RMSE
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

6.特征选择

我们从前面的图形中发现，newspaper的广告投入与销售量的线性关系并不强，因此我们可以猜测，newspaper的广告对销售量预测的帮助可能不大，从特征中移除可能对提升型效果有帮助。

```python
# 6.特征选择，剔除掉newspaper特征
feature_cols = ['TV', 'radio']
X = data[feature_cols]
y = data.sales
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
linreg = LinearRegression()
linreg.fit(X_train, y_train)
y_pred = linreg.predict(X_test)
np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```
